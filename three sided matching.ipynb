```python
def is_stable_my(first_set, second_set, third_set, matching):
    for pair in matching:
        third = matching[pair]
        first, second = pair
        # checking deviation for first element of matching
        Y_choices = first_set[first][0][:first_set[first][0].index(second)+1]
        Z_choices = first_set[first][1][:first_set[first][1].index(third)]
        for Y in Y_choices:
            for Z in Z_choices:
                Y_match = [(key[0],matching[key]) for key in matching if key[1]==Y][0]
                
                if second_set[Y][0].index(Z)>second_set[Y][0].index(Y_match[1]):
                    continue
                elif (second_set[Y][0].index(Z)==second_set[Y][0].index(Y_match[1]) and second_set[Y][1].index(first)>second_set[Y][1].index(Y_match[0])):
                    continue
                
                Z_match = [key for key,val in matching.items() if val==Z][0]
                if third_set[Z][0].index(first)<third_set[Z][0].index(Z_match[0]):
#                     logging.info("%s-",datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
                    logging.info("%s, %s, %s %s",first, Y, Z,"can form a coalition and block the matching!")
                    return False
                
                elif (third_set[Z][0].index(first)==third_set[Z][0].index(Z_match[0]) and third_set[Z][1].index(Y)<third_set[Z][1].index(Z_match[1])):
#                     logging.info("%s-",datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
                    logging.info("%s, %s, %s %s",first, Y, Z,"can form a coalition and block the matching!")
                    return False
#         print("No deviation found for",first)
        # checking deviation for second element of matching
        Z_choices = second_set[second][0][:second_set[second][0].index(third)+1]
        X_choices = second_set[second][1][:second_set[second][1].index(first)]
        for Z in Z_choices:
            for X in X_choices:
                
                Z_match = [key for key,val in matching.items() if val==Z][0]
                if third_set[Z][0].index(X)>third_set[Z][0].index(Z_match[0]):
                    continue
                elif (third_set[Z][0].index(X)==third_set[Z][0].index(Z_match[0]) and third_set[Z][1].index(second)>third_set[Z][1].index(Z_match[1])):
                    continue
                X_match = [(key[1],matching[key]) for key in matching if key[0]==X][0]
                if first_set[X][0].index(second)<first_set[X][0].index(X_match[0]):
#                     logging.info("%s-",datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
                    logging.info("%s, %s, %s %s",X, second, Z,"can form a coalition and block the matching!")
                    return False
                
                elif (first_set[X][0].index(second)==first_set[X][0].index(X_match[0]) and first_set[X][1].index(Z)<first_set[X][1].index(X_match[1])):
#                     logging.info("%s-",datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
                    logging.info("%s, %s, %s %s",X, second, Z,"can form a coalition and block the matching!")
                    return False
#         print("No deviation found for",second)     
    
        # checking deviation for third element of matching
        X_choices = third_set[third][0][:third_set[third][0].index(first)+1]
        Y_choices = third_set[third][1][:third_set[third][1].index(second)]
        for X in X_choices:
            for Y in Y_choices:
                
                X_match = [(key[1],matching[key]) for key in matching if key[0]==X][0]
                if first_set[X][0].index(Y)>first_set[X][0].index(X_match[0]):
                    continue
                elif (first_set[X][0].index(Y)==first_set[X][0].index(X_match[0]) and first_set[X][1].index(third)>first_set[X][1].index(X_match[1])):
                    continue
                Y_match = [(key[0],matching[key]) for key in matching if key[1]==Y][0]
                if second_set[Y][0].index(third)<second_set[Y][0].index(Y_match[1]):
#                     logging.info("%s-",datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
                    logging.info("%s, %s, %s %s",X, Y, third,"can form a coalition and block the matching!")
                    return False
                
                elif (second_set[Y][0].index(third)==second_set[Y][0].index(Y_match[1]) and second_set[Y][1].index(X)<second_set[Y][1].index(Y_match[0])):
#                     logging.info("%s-",datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
                    logging.info("%s, %s, %s %s",X, Y, third,"can form a coalition and block the matching!")
                    return False
#         print("No deviation found for",third)
    return True
```
